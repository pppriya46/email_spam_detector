```python
import pandas as pd
from sklearn import metrics
```

```python
bench = pd.read_csv('Benchmark Results.csv')
bench
```

```python
bench['Classification'] = bench.Classification.map({'not spam':0, 'spam':1})
bench
```

```python
label = bench['Label']
classification = bench['Classification']
```

```python
print(metrics.classification_report(label, classification))
```

```python
pd.DataFrame(
    metrics.confusion_matrix(label, classification),
    index = [['actual', 'actual'], ['ham', 'spam']],
    columns = [['predicted', 'predicted'], ['ham', 'spam']]
    )
```
